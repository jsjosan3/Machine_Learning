```python
x=df.drop('TARGET CLASS', axis=1)
x.head()
```

```python
y=df['TARGET CLASS']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler  
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train=pd.DataFrame(X_train,columns= x.columns)
```

```python
X_train
```

```python
X_test=pd.DataFrame(X_test,columns=x.columns)
X_test
```

```python
 from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clas=KNeighborsClassifier(n_neighbors=1)
```

```python
knn_clas.fit(X_train,y_train)
```

```python
knn_clas.score(X_test,y_test)
```

```python
y_pred=knn_clas.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
conf_mat=confusion_matrix(y_pred,y_test)
conf_mat
```

```python
error_rate=[]
for i in range (1,50):
    knn_clas=KNeighborsClassifier(n_neighbors=i)
    knn_clas.fit(X_train,y_train)
    y_pred_i=knn_clas.predict(X_test)
    error_rate.append(np.mean(y_pred_i != y_test))
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1,50),error_rate,linestyle='--',color="red",marker='*')
plt.xlabel('value of k',color="blue",size=15)
plt.ylabel('Error Rate',color="blue",size=15)
plt.title('Error Rate vs K value', color="red")
plt.show()
```

```python
knn_clas=KNeighborsClassifier(n_neighbors=19)
knn_clas.fit(X_train,y_train)
y_pred=knn_clas.predict(X_test)
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
cm=(confusion_matrix(y_pred,y_test))
print(cm)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
plt.show()
```
